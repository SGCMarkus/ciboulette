## Exemple WebObs AAVSO curve 

- Import class webobs

```python
from ciboulette.aavso.webobs import WebObs
```

- Create curve with visual filter on WebObs page

```python
c1405 = WebObs(nameID='v1405 cas',filtername='vis')
```

- Plot curve

```python
c1405.plot()
```

- Reading observations (only the first page)

```python
c1405.observations
```

- Create curve with visual filter on WebObs page

```python
c = WebObs(nameID='v1674 her',filtername='vis')
```

- Plot curve

```python
c.plot()
```

- Reading comments line and period

```python
c.comments, c.period
```

- Mask to modify a value, show magnitudes less than 6.8

```python
mask = c.observations['Magnitude']<6.8
c.observation[mask]
```

- Modify the value of the search JD

```python
y = 0
for i in c.observations:
    if i['JD'] == 2459388.43056:
        index = y
        c.observations['Magnitude'][index]=11        
    y = y+1    
```

```python
c.plot()
```

- Search for an observer

```python
mask = c.observations['Observer'] == 'EBAA'
c.observation[mask]
```

- Filter the display of observations

```python
data = c.observations
data['JD','Magnitude','Filter','Observer','Notes'].pprint(max_width=200,max_lines=255)
```
